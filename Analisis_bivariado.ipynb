```python
# Replace '-' with NaN and convert columns to numeric type
for col in df_columnas_a_comparar.columns:
    df_columnas_a_comparar[col] = pd.to_numeric(df_columnas_a_comparar[col].replace('-', np.nan))

df_columnas_a_comparar.corr()
```

```python
df_columnas_a_comparar.corr()
```

```python
df_columnas_a_comparar.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(df_columnas_a_comparar.corr(), annot=True, cmap='coolwarm')
plt.show()
```


El análisis bivariado realizad muestra una correlación positiva entre los diferentes percentiles de las pruebas ICFES en Lectura Crítica, Matemáticas, Ciencias Naturales y Sociales Ciudadanas, con valores que oscilan entre 0.72 y 0.8. Esto indica que un buen rendimiento en una de estas áreas generalmente se asocia con un desempeño sólido en las demás. La correlación más fuerte, de 0.8, se observa entre Matemáticas y Ciencias Naturales, lo que sugiere una interrelación significativa entre estas áreas. Esta información es útil para enfocar estrategias de mejora educativa de manera integral.
